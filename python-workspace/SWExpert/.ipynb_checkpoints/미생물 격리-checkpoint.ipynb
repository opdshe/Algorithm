```python
class virus:
    def __init__(self, y, x, num, dir):
        self.y=y
        self.x=x
        self. num=num
        self.dir=dir

def move(virus):
    #(: 1, : 2, : 3, : 4)
    ny, nx =None, None
    if virus.dir==1:
        ny, nx = virus.y-1, virus.x
    elif virus.dir==2:
        ny, nx = virus.y+1, virus.x
    elif virus.dir==3:
        ny, nx = virus.y, virus.x-1
    elif virus.dir==4:
        ny, nx = virus.y, virus.x+1

    virus.y, virus.x= ny, nx
    if board[ny][nx]==1:
        if virus.dir==1:
            virus.dir=2
        elif virus.dir==2:
            virus.dir=1
        elif virus.dir==3:
            virus.dir=4
        elif virus.dir==4:
            virus.dir=3
        virus.num=int(virus.num/2)

def operate():
    global viruses, M
    coordSet=set()
    checkList={}
    while M:
        #            ,   . 
        for key, virus in viruses.items():
            move(virus)
        #    
        for key, virus in viruses.items():
            if (virus.y, virus.x) in checkList:
                checkList[(virus.y, virus.x)].append(key)
            else:
                checkList[(virus.y, virus.x)]=[key]
        temp=[]
        for key, list in checkList.items():
            if len(list)<=1:
                temp.append(key)
        for idx in temp:
            del checkList[idx]
        #checkList  idx   
        for key, val in checkList.items():
            maxIdx=-1
            maxVal=-1

            for node in val:
                if maxVal<viruses[node].num:
                    maxVal=viruses[node].num
                    maxIdx=node
            subList=[]
            for node in val:
                if node!=maxIdx:
                    viruses[maxIdx].num+=viruses[node].num
                    subList.append(node)
            for sub in subList:
                del viruses[sub]
        M-=1
        checkList={}

    sum=0
    for key, val in viruses.items():
        sum+=val.num
        #print("{}  {}{} {}".format(key, val.y, val.x, val.num))
    print("#{} {}".format(test_case, sum))
            
            
T = int(input())
#    ,  .
for test_case in range(1, T + 1):
    N, M, K = map(int, input().split())
    viruses={}

    for i in range(K):
        y, x, num, dir=map(int, input().split())
        viruses[i]=(virus(y, x, num, dir))
    board=[[0]*N for _ in range(N)]
    for i in range(N):
        for j in range(N):
            if i==0 or j==0 or i==N-1 or j==N-1:
                board[i][j]=1
    operate()
```
